```python
def value_imputation(dataset, lower_outlier, upper_outlier, columns):
    # Iterate through each column
    for col in columns:
        # Check if the column is numeric
        if dataset[col].dtype == 'float64' or dataset[col].dtype == 'int64':
            # Check if there are any missing values or outliers
            if dataset[col].isnull().any() or ((dataset[col] > upper_outlier) | (dataset[col] < lower_outlier)).any():
                # Replace outliers with median
                dataset[col] = dataset[col].apply(lambda x: dataset[col].median() if (x > upper_outlier or x < lower_outlier) else x)
                # Fill missing values with mean
                if dataset[col].isnull().any():
                    dataset[col].fillna(dataset[col].mean(), inplace=True)
        else:
            # Print a message for categorical columns
            print("It's a Categorical Column. Please Enter Numerical Column")
    return dataset

# Load the dataset
data = pd.read_csv(r'C:\Users\pv437\Desktop\Data Scince Folder\Linked in\Data Science Datasets\marks.csv')

# Define columns to process
columns = ['math_score', 'reading_score', 'writing_score']

# Apply value imputation
# lower_outlier_value = Minimum Value (Code will remove Value Above that Value)
# Upper_outlier_value = Maximum Value (Code will remove Value Below that Value)
Filled_Data = value_imputation(data, 0, 100, columns)
```
